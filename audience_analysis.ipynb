```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 1000)
```

Individual Level

```python
session_data = pd.read_excel('session_data.xlsx')
session_data
```

```python
print('Mean Bounce Rate was ', np.mean(session_data['Bounce Rate']))
print('Mean Avg. Session Duration was ', np.mean(session_data['Avg. Session Duration']))
print('Max Avg. Session Duration was ', np.max(session_data['Avg. Session Duration']))
print('Min Avg. Session Duration was ', np.min(session_data['Avg. Session Duration']))
print('Mean Revenue was ', np.mean(session_data['Revenue']))
print('Max Revenue was ', np.max(session_data['Revenue']))
print('Min Revenue was ', np.min(session_data['Revenue']))
print('Max Transcations was ', np.max(session_data['Transactions']))
print('Mean Transcations was ', np.mean(session_data['Transactions']))
print('Mean Goal Conversion Rate was ', np.mean(session_data['Goal Conversion Rate']))
print('Max Goal Conversion Rate was ', np.max(session_data['Goal Conversion Rate']))
print('Mean Sessions was ', np.mean(session_data['Sessions']))
print('Max Sessions was ', np.max(session_data['Sessions']))
```

```python
import matplotlib.pyplot as plt
plt.plot(session_data['Sessions'], session_data['Avg. Session Duration'], marker='o', linestyle='-', color='b')
plt.xlabel('Sessions')
plt.ylabel('Avg. Session Duration')
plt.title('Line Plot')
plt.show()
```

```python
len(session_data[session_data['Avg. Session Duration'] < 10]), len(session_data)
```

Majority of the people who spend large amount of time have less sessions. Looks like they know what they want and website helps them find it.

```python
import matplotlib.pyplot as plt
plt.plot(session_data['Revenue'], session_data['Avg. Session Duration'], marker='o', linestyle=None, color='g')
plt.xlabel('Revenue')
plt.ylabel('Avg. Session Duration')
plt.title('Line Plot')
plt.show()
```

Product Data

```python
product_data = pd.read_excel('product_data.xlsx')
product_data
```

```python
product_sales = product_data.groupby('Product')['Product Revenue'].sum().reset_index()

top_selling_products = product_sales.sort_values('Product Revenue', ascending=False)
print(top_selling_products)
top_selling_products.to_excel('top_products.xlsx')
```

```python
product_purchases = product_data.groupby('Product')['Quantity'].sum().reset_index()
top_quantity_products = product_purchases.sort_values('Quantity', ascending=False)
print(top_quantity_products)
top_quantity_products.to_excel('top_quantity.xlsx')
```
